Task 1: Improve the image quality by applying brightness and contrast enhancement techniques.

Ans:
 Apply Contrast Limited Adaptive Histogram Equalization (CLAHE),AHE,GHE among them CLAHE presents better result

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('/content/2.jpg', cv2.IMREAD_GRAYSCALE)



clahe = cv2.createCLAHE(clipLimit=4.0, tileGridSize=(4, 6))
clahe_image = clahe.apply(image)

# Plotting the images
plt.figure(figsize=(8,5))

plt.subplot(1, 4, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(clahe_image, cmap='gray')
plt.title('Contrast Limited AHE (CLAHE)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

Task 2:Segment facial features to isolate eyes, mouth, and other key components.

Ans:
**I have used morphological operation,Global Thresholding, Adaptive mean and Adaptive Gaussian Thresholding method. Among them Adaptive mean provides better result i think.**

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('/content/1.jpg', cv.IMREAD_GRAYSCALE)



th1 = cv.adaptiveThreshold(img,255,cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY,13,2)

plt.figure(figsize=(6,4))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt
plt.subplot(1, 2, 2)
plt.imshow(th1, cmap='gray')
plt.title('Adaptive Mean Thresholding')
plt.axis('off')

plt.tight_layout()

plt.show()
```

Task 3: Detect edges to analyze facial structures and expressions.

Answer:
I have examined with vertical,horizontal,combined them, laplacian,canny and sobel filter. Among them canny Image make a good result of detecting images edge. **bold text**

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img=cv.imread('/content/4.jpg', cv.IMREAD_GRAYSCALE)

sobel_x=cv.Sobel(img,-1,1,0,3)#changes in y axis ;(src_img,ddepth,dx,dy,kernel_size); 0 means actual value
sobel_y=cv.Sobel(img,-1,0,1,3)#changes in x axis

lower_threshold=150
upper_threshold=250
canny=cv.Canny(img,lower_threshold,upper_threshold)

plt.figure(figsize=(8,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(canny, cmap='gray')
plt.title('Canny filter')
plt.axis('off')


plt.show()
```

Task 4: Use image blurring to soften non-essential features while keeping the focus on facial expressions.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/2.jpg')
#img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
assert img is not None, "file could not be read, check with os.path.exists()"

bilateral_blur = cv.bilateralFilter(img, 10, 20, 30)


plt.figure(figsize=(8, 5))

plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')


plt.subplot(122)
plt.imshow(bilateral_blur, cmap='gray')
plt.title('Bilateral')
plt.axis('off')

plt.show()
```
